```python
df = pd.read_csv('../data/lyricskaggle.csv')
model = SentenceTransformer('all-distilroberta-v1')
lyrics = df['lyrics']
model2 = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
embeddings2 = model.encode(lyrics, batch_size=200, show_progress_bar=True, convert_to_tensor=True)
torch.save(embeddings2, "384_embeddings.pt")
#embeddings2 = pd.DataFrame(embeddings2)
#embeddings2.to_csv('384_embeddings.csv')
```

```python
embeddings_list = embeddings2.detach().numpy()
print(embeddings_list)
```

```python
df.head()
```

```python
gender = df['gender']
genre = df['genre']

tensor_dict = {}
count_dict = {}

for i in range(len(gender)):
    key = gender[i] + "" + genre[i]
    if not key in tensor_dict:
        tensor_dict[key] = embeddings_list[i]
        count_dict[key] = 1
    else:
        tensor_dict[key] += embeddings_list[i]
        count_dict[key] += 1

print(tensor_dict)
print(count_dict)
```

```python
for key in tensor_dict.keys():
    tensor_dict[key] = tensor_dict[key]/count_dict[key]

print(tensor_dict)
```
